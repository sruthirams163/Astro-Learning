```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.timeseries import LombScargle
import pandas as pd

df = pd.read_csv("51_peg_self.csv")


# Load the data
t = df['JD'].values
rv = df['Radial_Velocity'].values
rv_err = df['Radial_Velocity_Uncertainty'].values

# Run Lomb-Scargle
frequency, power = LombScargle(t, rv, rv_err).autopower(minimum_frequency=0.1, 
                                                          maximum_frequency=2.0)

# Convert frequency to period in days
period = 1.0 / frequency

# Find the peak
best_frequency = frequency[np.argmax(power)]
best_period = 1.0 / best_frequency

print(f"Best period found: {best_period:.4f} days")
print(f"Expected from thesis: 4.2308 days")

# Plot
plt.figure(figsize=(10, 4))
plt.plot(period, power, color='steelblue', linewidth=0.8)
plt.axvline(best_period, color='red', linestyle='--', label=f'Peak: {best_period:.3f} days')
plt.xlabel('Period (days)')
plt.ylabel('Lomb-Scargle Power')
plt.title('Lomb-Scargle Periodogram — 51 Pegasi b self')
plt.xscale('log')
plt.xlim(0.5, 10)
plt.legend()
plt.tight_layout()
plt.savefig('51peg_self_periodogram.png', dpi=150)
plt.show()
print("Done")
```

```python
period = 4.2310

# Phase fold: this compresses all observations onto one single orbit
# The % operator gives the remainder — so any time value maps to 0 to period
# Dividing by period maps it to 0 to 1 (one full orbit = phase 0 to 1)
phase = (t % period) / period

# Plot the phase-folded curve
fig, ax = plt.subplots(figsize=(8, 5))

ax.errorbar(phase, rv, yerr=rv_err,
            fmt='o',           # circular markers
            color='steelblue',
            ecolor='lightgray',  # error bar color
            elinewidth=1,
            capsize=3,
            markersize=4,
            label='Observed RV')

ax.axhline(0, color='gray', linewidth=0.5, linestyle='--')  # zero line
ax.set_xlabel('Phase (time / period)', fontsize=12)
ax.set_ylabel('Radial Velocity (m/s)', fontsize=12)
ax.set_title('Phase-folded RV curve — 51 Pegasi b self', fontsize=13)
ax.legend()
plt.tight_layout()
plt.savefig("51peg_self_phase_folded.png", dpi=150)
plt.show()
print("Phase fold done.")
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

import os
os.makedirs("51_peg_self", exist_ok=True)

# Load your saved phase-folded data
# (paste your phase and rv arrays from the previous cell here)
# For example, if your phase-fold cell produced variables called 'phase' and 'rv':
# phase = ... (your phase array)
# rv = ...    (your RV array)
# If you ran it in the same notebook, they're already in memory - just continue below.

# Step 1: Define the Keplerian RV function
# This is exactly Equation 2.8 from your thesis:
# vr = K * [cos(theta + omega) + e * cos(omega)]
# For a circular orbit (e=0), this simplifies to a pure cosine.
# We fit phase (0 to 1) instead of true anomaly, so we use 2*pi*phase.

def keplerian_rv(phase, K, e, omega, v_sys):
    """
    phase  : orbital phase from 0 to 1
    K      : semi-amplitude in m/s
    e      : eccentricity (0 = circular)
    omega  : argument of periastron in radians
    v_sys  : systemic velocity (shifts whole curve up or down)
    """
    theta = 2 * np.pi * phase          # convert phase to angle in radians
    vr = K * (np.cos(theta + omega) + e * np.cos(omega)) + v_sys
    return vr

# Step 2: Give the fitter a starting guess for each parameter
# K ~ 56 m/s  (from your thesis result)
# e ~ 0       (nearly circular orbit)
# omega ~ 4.7 (from your thesis result, in radians)
# v_sys ~ 0   (we don't know the systemic velocity, start at 0)

p0 = [56, 0.01, 4.7, 0]

# Step 3: Run the fit
# curve_fit returns:
#   popt = best-fit parameter values
#   pcov = covariance matrix (we use diagonal to get uncertainties)

popt, pcov = curve_fit(keplerian_rv, phase, rv, p0=p0)

# Step 4: Extract results
K_fit, e_fit, omega_fit, vsys_fit = popt
perr = np.sqrt(np.diag(pcov))   # 1-sigma uncertainties on each parameter

print(f"Best-fit K     : {K_fit:.2f} ± {perr[0]:.2f} m/s   (thesis: 55.94 m/s)")
print(f"Best-fit e     : {e_fit:.4f} ± {perr[1]:.4f}        (thesis: 0.013)")
print(f"Best-fit omega : {omega_fit:.4f} ± {perr[2]:.4f} rad")
print(f"Systemic vel   : {vsys_fit:.2f} ± {perr[3]:.2f} m/s")

# Step 5: Plot the fit over your phase-folded data
phase_model = np.linspace(0, 1, 1000)   # smooth phase array for the model curve
rv_model = keplerian_rv(phase_model, *popt)

plt.figure(figsize=(8, 5))
plt.scatter(phase, rv, s=20, color='steelblue', label='Observed RV', zorder=3)
plt.plot(phase_model, rv_model, color='crimson', lw=2, label='Keplerian fit')
plt.xlabel("Phase (time / period)")
plt.ylabel("Radial Velocity (m/s)")
plt.title("51 Pegasi b self — Keplerian model fit")
plt.legend()
plt.tight_layout()
plt.savefig("51_peg_self/51peg_self_keplerian_fit.png", dpi=150)
plt.show()
print("Plot saved.")
```

```python
# Constants
G = 6.674e-11          # gravitational constant in SI units
M_sun = 1.989e30       # solar mass in kg
M_jup = 1.898e27       # Jupiter mass in kg

# Known values
M_star = 1.03 * M_sun  # 51 Peg stellar mass
P = 4.2310 * 86400     # orbital period converted from days to seconds
K = K_fit              # your best-fit semi-amplitude in m/s

# Equation 2.12 from your thesis — minimum planetary mass
mp_sini = (P / (2 * np.pi * G))**(1/3) * K * M_star**(2/3)

print(f"Minimum mass mp*sin(i) : {mp_sini / M_jup:.4f} Jupiter masses")
print(f"Thesis value           : 0.4757 Jupiter masses")
```

## Results and comparison

| Parameter | This analysis | Thesis value | Literature |
|-----------|--------------|--------------|------------|
| Period (days) | 4.2310 | 4.2308 | 4.230785 |
| K (m/s) | 56.10 ± 0.73 | 55.83 | 55.94 ± 0.69 |
| mp sin(i) (MJup) | 0.4555 | 0.4757 | 0.472 |

The ~4% difference in minimum mass arises from fixing e=0 in the 
circular fit. The real orbit has e=0.013, which slightly affects 
the fitted K and propagates into the mass estimate.
